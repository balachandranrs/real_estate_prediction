## Global Imports

```python
import pandas
import hdbscan
from sklearn.preprocessing import StandardScaler
```

## HDB Clustering

```python
clusDataOriginal = pandas.read_csv("/home/vivek/Documents/mydata/iimb/clusteting/clustering_dataset.csv",skiprows=[1])
haversine = pandas.read_csv("/home/vivek/Documents/mydata/iimb/satellite_images/haversine.csv")
```

```python
clusDataOriginal = clusDataOriginal.drop(["latitude","longitude"],axis=1)
clusData = haversine.merge(clusDataOriginal,how="left",on="post_code")
post_codes = list(clusData.post_code)
clusData = clusData.drop(["post_code"],axis=1)
```

```python
scaler = StandardScaler()
priceCols = ['price']
clusDataFiltered = clusData[priceCols]
clusDataFiltered = scaler.fit_transform( clusDataFiltered[priceCols])
```

### Trying Different Parameters To Find Stable Clusters

```python
for j in [1000,5000,10000,20000]:
    for i in range(1,15):
        clusterer = hdbscan.HDBSCAN(min_cluster_size=j,min_samples=i)
        clusterer.fit(clusDataFiltered)
        clusters = list(clusterer.labels_)
        clusSize = len(set(clusters))
        print ((j,i, clusSize)) 
```

### Running HDB With Optimum Parameters

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=5000,min_samples=13)
clusterer.fit(clusDataFiltered)
clusters = clusterer.labels_
clusters_df = pandas.DataFrame( { "cluster":clusters, "post_code":post_codes,"mean_price":clusData.price  } )
clusters_df.to_csv("/home/vivek/Documents/mydata/iimb/clusteting/cluster_hdb_output.csv",index=False)
clusters_df.cluster.value_counts()
```
